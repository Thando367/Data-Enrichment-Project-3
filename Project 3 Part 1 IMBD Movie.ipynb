```python
import pandas as pd
import numpy as np

# basics data
basics_url = "https://datasets.imdbws.com/title.basics.tsv.gz"
```

```python
# akas data
akas_url = "https://datasets.imdbws.com/title.akas.tsv.gz"
```

```python
# ratings data
ratings_url = 'https://datasets.imdbws.com/title.ratings.tsv.gz'
```

```python
# importing the data
```

```python
basics = pd.read_csv(basics_url, sep='\t', low_memory=False)
```

```python
akas = pd.read_csv(akas_url, sep='\t', low_memory=False)
```

```python
ratings = pd.read_csv(ratings_url, sep='\t', low_memory=False)
```

```python
# CLEANING the DATA
```

```python
# replacing \N values with NaN
basics.replace({'\\N':np.nan}, inplace=True)
```

```python
# eliminate any movies that are null for runtimeMinutes and genre
basics.dropna(subset=['runtimeMinutes', 'genres'], inplace=True)
```

```python
basics = basics[basics['titleType']=='movie']
```

```python
# keep startYear 2000-2022
basics = basics[(basics['startYear'] >= '2000') & (basics['startYear'] <= '2022')]
```

```python
is_documentary = basics['genres'].str.contains('documentary', case=False)
basics =basics[~is_documentary]
```

```python
# include US by using the filter akas dataframe
basics['tconst'].isin(akas['titleId'])
```

```python
basics.head()
```

```python
# akas data
# replacing \N values with NaN
akas.replace({'\\N':np.nan}, inplace=True)
```

```python
akas = akas[akas['region'] == 'US']
```

```python
akas.head()
```

```python
# ratings data
# replacing \N values with NaN
ratings.replace({'\\N':np.nan}, inplace=True)
```

```python
# keep only US movies
ratings = ratings[ratings['tconst'].isin(akas['titleId'])]
```
