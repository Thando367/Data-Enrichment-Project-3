```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# LOADING DATA
year_2000 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
year_2001 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
```

```python
year_2001.info()
```

```python
df = pd.concat([year_2000, year_2001], ignore_index=True)
df.head
```

```python
movie_finance = df[(df['budget'] > 0) | (df['revenue'] > 0)]
movie_finance.head(2)
```

```python
# Number of movies with valid financial information
movie_finance = len(movie_finance)
print(f"Number of movies with valid financial information (Budget or Revenue): {movie_finance}")
```

```python
# Additional EDA or visualizations with valid financial information
plt.figure(figsize=(10, 6))
sns.scatterplot(x='budget', y='revenue', data=movie_finance)
plt.title('Scatter Plot of Budget vs. Revenue')
plt.xlabel('budget')
plt.ylabel('revenue')
plt.show()
```

```python
# Count the number of movies in each certification category
certification_counts = movie_finance['certification'].value_counts()
certification_counts# Display the counts
```

```python
# Bar plot for certification counts
certification_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Movies in Each Certification Category')
plt.xlabel('Certification')
plt.ylabel('Number of Movies')
plt.show()
```

```python
# Group by certification and calculate the average revenue for each category
average_revenue_per_certification = movie_finance.groupby('certification')['revenue'].mean()
print(average_revenue_per_certification)
```

```python
# Set the display format to show numbers without scientific notation
pd.options.display.float_format = '{:,.2f}'.format
```

```python
print(average_revenue_per_certification)
```

```python
# Bar plot for average revenue per certification category
average_revenue_per_certification.plot(kind='bar', color='coral')
plt.title('Average Revenue per Certification Category')
plt.xlabel('Certification Category')
plt.ylabel('Average Revenue')
plt.show()
```
